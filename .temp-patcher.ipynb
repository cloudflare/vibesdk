```python
from pathlib import Path
import re

file_path = Path('/workspaces/vibesdk/src/components/layout/global-header.tsx')
text = file_path.read_text()

# Remove CloudflareLogo import if still present
text = re.sub(r"import \{ CloudflareLogo \} from '../icons/logos';\n", '', text)

# Clean up the broken RIP ViBE SDK block
pattern = re.compile(
    r'(<SidebarTrigger[^>]+/>\s*)'
    r'<div className="ml-3 text-sm font-semibold text-text-primary">RIP ViBE SDK</div>\s*'
    r'/>\s*',
    re.DOTALL,
)
replacement = (
    r"\1\t\t\t\t\t<div className=\"ml-3 text-sm font-semibold text-text-primary\">RIP ViBE SDK</div>\n"
)
text, count = pattern.subn(replacement, text)

if count == 0:
    raise RuntimeError('Did not find the broken RIP ViBE SDK block to patch in global-header.tsx')

file_path.write_text(text)
print('Patched global-header.tsx, replacements made:', count)
```
